#### Define
Check if the dogs ratings are provided correctly

```python
# check dogs rating_numerator
import re
def rating_numerator(text_string):
    rating_pattern=re.compile('(\d+)\/(\d+)')
    result=rating_pattern.findall(text_string)
    if result[0][1]=='10':
        return result[0][0]
    
```
